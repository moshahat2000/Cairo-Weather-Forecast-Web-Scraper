```python
import pandas as pd
from datetime import datetime
import requests
from bs4 import BeautifulSoup
```

```python
link = requests.get("https://www.timeanddate.com/weather/egypt/cairo/ext")
link.status_code
```

```python
soup = BeautifulSoup(link.text, 'html.parser')
soup
```

```python
all = soup.find_all('th')
all
```

```python
date = []
for th in all:
    span = th.find('span', class_= 'smaller soft')
    if span:
        date.append(th)
date
```

```python
days = []
dates = []
for th in date:
    parts = th.get_text(separator= " ", strip= True).split()
    days.append(parts[0])
    dates.append(parts[1:])
```

```python
days, dates
```

```python
dates_text = []
for d in dates:
    day_month = ' '.join(d)
    dates_text.append(day_month)
dates_text
```

```python
dates_fixed = []
for d in dates_text:
    fixed_d = datetime.strptime(d, "%d %b")
    dates_fixed.append(fixed_d)
dates_fixed
```

```python
dates_str = [dt.strftime("%d %b") for dt in dates_fixed]
dates_str
```

```python
all2 = soup.find_all('td')
all2
```

```python
temp = []
for td in all2:
    temp_text = td.get_text(strip= True)
    if '/' in temp_text and '°C' in temp_text:
        temp_text = temp_text.replace('\xa0', ' ')
        temp.append(temp_text)

temp 
```

```python
weather = []
for td in all2:
    if 'class' in td.attrs and 'small' in td['class']:
        weather_text = td.get_text(strip= True).replace('.', '')
        weather.append(weather_text)

weather
```

```python
df = pd.DataFrame({'day': days, 'date': dates_str, 'temperature': temp, 'weather': weather})
df
```
